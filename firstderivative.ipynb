```python
import numpy as n 
from math import * 
import matplotlib.pyplot as p
```

```python
xmax=eval(input("Domain "))
nx=eval(input("Sample Points"))
dx=xmax/(nx-1)
x=n.linspace(-xmax,xmax,nx)
l=20*dx
f=e**(-l*(x**2))
```

```python
p.figure(figsize=(5,5))
p.plot(x,f)
p.title('Gauss Boi')
p.xlabel('x')
p.ylabel('f(x)')
p.grid()
p.show()
```

```python
#analytical derivative
ad=-2*x*f
ad[0],ad[nx-1]=0,0
p.plot(x,ad)
p.grid()
p.show()
```

```python
#numerical derivative
nd=n.zeros(nx)
for i in range(1,nx-1):
    nd[i]=(f[i+1]-f[i-1])/2*dx
p.plot(x,nd)
p.grid()
p.show()
```

```python
rms=n.sqrt(n.mean((ad-nd)**2))
print(rms)
```

```python
ll=n.zeros(200)
err=n.zeros(200)
nd=n.zeros(200)
t=-1
for i in range(1,201):
    t=t+1
    ll[t]=i
    k=i*dx
    
    f0=e**(-k*(x**2))
    for j in range(1,199):
        nd[j]=(f0[j+1]-f0[j-1])/2*dx
    ad=-2*x*(e**(-k*(x**2)))
    ad[0],ad[199]=0,0
    err[t]=n.sqrt(n.mean((nd-ad)**2))
    
```

```python
p.plot(ll,err)
p.grid()
p.show()
```
